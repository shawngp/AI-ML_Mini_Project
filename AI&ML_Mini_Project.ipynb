# K-Nearest Neighbours

```python
import numpy as np
url=(r'https://github.com/srafay/Machine_Learning_A-Z/raw/master/Part%203%20-%20Classification/Section%2014%20-%20Logistic%20Regression/Social_Network_Ads.csv')
df = pd.read_csv(url)
X = df.iloc[:, [2, 3]].values
y = df.iloc[:, 4].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Fitting K-NN to the Training set
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5)
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# create heatmap
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(7,5))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
#print classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

# K-Means Clustering

```python
# import all the required packages
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
url=(r'https://github.com/srafay/Machine_Learning_A-Z/raw/master/Part%203%20-%20Classification/Section%2014%20-%20Logistic%20Regression/Social_Network_Ads.csv')
df = pd.read_csv(url)
df.head()
```

```python
# visualize the dataset using scatter plot
plt.scatter(df.Age,df['EstimatedSalary'])
plt.xlabel('Age')
plt.ylabel('EstimatedSalary')
```

```python
# create clusters for random value of k=3
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age','EstimatedSalary']])
y_predicted
df['cluster']=y_predicted
df.head()
```

```python
# analyze cluster centers
km.cluster_centers_
```

```python
# draw clusters with cluster center
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Age,df1['EstimatedSalary'],color='green')
plt.scatter(df2.Age,df2['EstimatedSalary'],color='red')
plt.scatter(df3.Age,df3['EstimatedSalary'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Age')
plt.ylabel('EstimatedSalary')
plt.legend()
```

```python
# data normalization and visualization
scaler = MinMaxScaler()
scaler.fit(df[['EstimatedSalary']])
df['EstimatedSalary'] = scaler.transform(df[['EstimatedSalary']])
scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
df.head()
plt.scatter(df.Age,df['EstimatedSalary'])
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age','EstimatedSalary']])
y_predicted
km.cluster_centers_
df['cluster']=y_predicted
df.head()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
```

```python
plt.scatter(df1.Age,df1['EstimatedSalary'],color='green')
plt.scatter(df2.Age,df2['EstimatedSalary'],color='red')
plt.scatter(df3.Age,df3['EstimatedSalary'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.legend()
```

```python
# calculate value of K using elbow method
sse = []
k_rng = range(1,10)
for k in k_rng:
 km = KMeans(n_clusters=k)
 km.fit(df[['Age','EstimatedSalary']])
 sse.append(km.inertia_)
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_predicted,km.labels_))
```

# Principal Component Analysis

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
url=(r'https://github.com/srafay/Machine_Learning_A-Z/raw/master/Part%203%20-%20Classification/Section%2014%20-%20Logistic%20Regression/Social_Network_Ads.csv')
df = pd.read_csv(url)
df.drop(['User ID','Gender'],axis='columns',inplace=True)
df.head()
```

```python
# create dataframe
# Import the sklearn function
from sklearn.preprocessing import StandardScaler
# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
X_scaled
df['target'] = df['Purchased']
y = df['target']
```

```python
# splitting the dataset and use logistic regression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=30)
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
# apply pca for retaining 95 percent useful features
from sklearn.decomposition import PCA
pca = PCA(0.95)
X_pca = pca.fit_transform(df)
X_pca.shape
```

```python
# split new dataframe and train the model
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```

```python
#from the above scenario we can see that despite dropping lots of features we obtainednear by similar accuracy
pca = PCA(n_components=2)
X_pca = pca.fit_transform(df)
X_pca.shape
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```
